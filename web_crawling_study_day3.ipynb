```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# requests.get()으로 url정보 요청하기
headers = {'User-Agent' : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.77 Safari/537.36"}

url = "https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop"

r = requests.get(url, headers = headers)
r.status_code
```

```python
# BeautifulSoup()으로 웹페이지 분석 준비하기
soup = BeautifulSoup(r.text, 'html.parser')  # r.content 대신 r.text도 가능
print(soup)
```

```python
# 선수 정보가 담긴 태그와 클래스 찾기
player_info = soup.find_all('tr', class_ = ['odd', 'even'])
player_info
```

```python
# 첫번쨰 요소 확인하기
print(player_info[0])
```

```python
# 전체 개수 확인하기
print(len(player_info))
```

```python
# 7개의 정보를 담을 빈 리스트 만들기 number, name, position, age, nation, team, value
number = []
name = []
position = []
age = []
nation = []
team = []
value = []
```

```python
# player_info에서 'td'태그만 모두 찾기
for info in player_info:
  player = info.find_all("td")
  print(player)
```

```python
print(player[0])
```

```python
# 해당 정보 찾아서 각 리스트에 .append로 추가하기
for info in player_info:
  player = info.find_all("td")
number.append(player[0].text)
print(number)
```
